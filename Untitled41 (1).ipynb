```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from glob import glob
from tqdm import tqdm

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Base path
base_path = "/content/drive/MyDrive/ml_mini_project"

# Correct folder names
train_img_path = os.path.join(base_path, "training_data")
train_mask_path = os.path.join(base_path, "training_ground_truth")  # ✅ fixed name
test_img_path = os.path.join(base_path, "test_data")
test_mask_path = os.path.join(base_path, "test_ground_truth")

# Check sample files
print("Training images:", os.listdir(train_img_path)[:5])
print("Training masks:", os.listdir(train_mask_path)[:5])

# Image size
IMG_HEIGHT = 128
IMG_WIDTH = 128

# Load and preprocess images
def load_images(image_dir, mask_dir, img_size=(IMG_HEIGHT, IMG_WIDTH)):
    image_files = sorted(glob(os.path.join(image_dir, "*.jpg")))
    mask_files = sorted(glob(os.path.join(mask_dir, "*.png")))

    if len(image_files) == 0 or len(mask_files) == 0:
        raise ValueError("No images or masks found. Check your paths and file extensions.")

    images = []
    masks = []

    for img_path, mask_path in tqdm(zip(image_files, mask_files), total=len(image_files)):
        img = cv2.imread(img_path)
        img = cv2.resize(img, img_size)
        img = img / 255.0  # Normalize

        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, img_size)
        mask = mask / 255.0
        mask = np.expand_dims(mask, axis=-1)

        images.append(img)
        masks.append(mask)

    return np.array(images), np.array(masks)

# Load training and test sets
X_train, Y_train = load_images(train_img_path, train_mask_path)
X_test, Y_test = load_images(test_img_path, test_mask_path)

# Confirm shapes
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, backend as K

# Dice coefficient and Dice loss
def dice_coefficient(y_true, y_pred, smooth=1):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1 - dice_coefficient(y_true, y_pred)

# U-Net model
def build_unet(input_shape=(128, 128, 3)):
    inputs = layers.Input(input_shape)

    # Encoder
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D()(c2)

    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(c3)
    p3 = layers.MaxPooling2D()(c3)

    # Bottleneck
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(p3)
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(c4)

    # Decoder
    u5 = layers.UpSampling2D()(c4)
    u5 = layers.concatenate([u5, c3])
    c5 = layers.Conv2D(64, 3, activation='relu', padding='same')(u5)
    c5 = layers.Conv2D(64, 3, activation='relu', padding='same')(c5)

    u6 = layers.UpSampling2D()(c5)
    u6 = layers.concatenate([u6, c2])
    c6 = layers.Conv2D(32, 3, activation='relu', padding='same')(u6)
    c6 = layers.Conv2D(32, 3, activation='relu', padding='same')(c6)

    u7 = layers.UpSampling2D()(c6)
    u7 = layers.concatenate([u7, c1])
    c7 = layers.Conv2D(16, 3, activation='relu', padding='same')(u7)
    c7 = layers.Conv2D(16, 3, activation='relu', padding='same')(c7)

    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c7)

    model = models.Model(inputs, outputs)
    return model

# Build and compile
model = build_unet()
model.compile(optimizer='adam', loss=dice_loss, metrics=[dice_coefficient, 'accuracy'])

model.summary()
```

```python
from tensorflow.keras.losses import Loss

class DiceLoss(Loss):
    def __init__(self, smooth=1):
        super().__init__()
        self.smooth = smooth

    def call(self, y_true, y_pred):
        y_true_f = K.flatten(y_true)
        y_pred_f = K.flatten(y_pred)
        intersection = K.sum(y_true_f * y_pred_f)
        return 1 - (2. * intersection + self.smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + self.smooth)
```

```python
from tensorflow.keras.layers import Conv2D, UpSampling2D, concatenate, Activation, Multiply, Add

def attention_block(x, g, inter_channels):
    theta_x = Conv2D(inter_channels, 1, strides=1, padding='same')(x)
    phi_g = Conv2D(inter_channels, 1, strides=1, padding='same')(g)
    add = Add()([theta_x, phi_g])
    act = Activation('relu')(add)
    psi = Conv2D(1, 1, strides=1, padding='same')(act)
    psi = Activation('sigmoid')(psi)
    return Multiply()([x, psi])
```

```python
def build_attention_unet(input_shape=(128, 128, 3)):
    inputs = layers.Input(input_shape)

    # Encoder
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D()(c2)

    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(c3)
    p3 = layers.MaxPooling2D()(c3)

    # Bottleneck
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(p3)
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(c4)

    # Decoder with attention
    u5 = UpSampling2D()(c4)
    a3 = attention_block(c3, u5, inter_channels=64)
    u5 = concatenate([u5, a3])
    c5 = layers.Conv2D(64, 3, activation='relu', padding='same')(u5)
    c5 = layers.Conv2D(64, 3, activation='relu', padding='same')(c5)

    u6 = UpSampling2D()(c5)
    a2 = attention_block(c2, u6, inter_channels=32)
    u6 = concatenate([u6, a2])
    c6 = layers.Conv2D(32, 3, activation='relu', padding='same')(u6)
    c6 = layers.Conv2D(32, 3, activation='relu', padding='same')(c6)

    u7 = UpSampling2D()(c6)
    a1 = attention_block(c1, u7, inter_channels=16)
    u7 = concatenate([u7, a1])
    c7 = layers.Conv2D(16, 3, activation='relu', padding='same')(u7)
    c7 = layers.Conv2D(16, 3, activation='relu', padding='same')(c7)

    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c7)

    return models.Model(inputs, outputs)
```

```python
class DropBlock2D(tf.keras.layers.Layer):
    def __init__(self, drop_prob=0.3, block_size=7):
        super(DropBlock2D, self).__init__()
        self.drop_prob = drop_prob
        self.block_size = block_size

    def call(self, x, training=None):
        if not training or self.drop_prob == 0.0:
            return x

        batch_size, height, width, channels = tf.shape(x)[0], tf.shape(x)[1], tf.shape(x)[2], tf.shape(x)[3]
        gamma = self.drop_prob * (height * width) / (self.block_size ** 2) / ((height - self.block_size + 1) * (width - self.block_size + 1))
        mask = tf.cast(tf.random.uniform((batch_size, height, width, channels)) < gamma, tf.float32)

        block_mask = tf.nn.max_pool2d(
            mask,
            ksize=self.block_size,
            strides=1,
            padding='SAME'
        )

        return x * (1 - block_mask)


```

```python
def build_unet_dropblock(input_shape=(128, 128, 3)):
    inputs = layers.Input(input_shape)

    # Encoder
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(inputs)
    c1 = DropBlock2D()(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(p1)
    c2 = DropBlock2D()(c2)
    p2 = layers.MaxPooling2D()(c2)

    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(p2)
    c3 = DropBlock2D()(c3)
    p3 = layers.MaxPooling2D()(c3)

    # Bottleneck
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(p3)
    c4 = DropBlock2D()(c4)

    # Decoder
    u5 = layers.UpSampling2D()(c4)
    u5 = layers.concatenate([u5, c3])
    c5 = layers.Conv2D(64, 3, activation='relu', padding='same')(u5)

    u6 = layers.UpSampling2D()(c5)
    u6 = layers.concatenate([u6, c2])
    c6 = layers.Conv2D(32, 3, activation='relu', padding='same')(u6)

    u7 = layers.UpSampling2D()(c6)
    u7 = layers.concatenate([u7, c1])
    c7 = layers.Conv2D(16, 3, activation='relu', padding='same')(u7)

    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c7)

    return models.Model(inputs, outputs)
```

```python
import gradio as gr
import numpy as np
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt

# Optional: Load model from file if saved
# model = tf.keras.models.load_model("/content/drive/MyDrive/ml_mini_project/unet_model.h5",
#                                    custom_objects={'dice_loss': dice_loss, 'dice_coefficient': dice_coefficient})

# If model is already built and trained in memory, skip loading

# Image size (must match training size)
IMG_HEIGHT = 128
IMG_WIDTH = 128

# Segmentation + overlay function
def segment_and_overlay(input_img):
    # Resize and normalize
    img = cv2.resize(input_img, (IMG_WIDTH, IMG_HEIGHT))
    img_norm = img / 255.0
    img_input = np.expand_dims(img_norm, axis=0)

    # Predict mask
    pred_mask = model.predict(img_input)[0]
    raw_mask = pred_mask.squeeze()

    # Threshold mask
    binary_mask = (raw_mask > 0.5).astype(np.uint8) * 255

    # Convert to 3-channel for overlay
    mask_rgb = cv2.cvtColor(binary_mask, cv2.COLOR_GRAY2RGB)

    # Overlay mask on original image
    overlay = cv2.addWeighted(img, 0.7, mask_rgb, 0.3, 0)

    return [img, binary_mask, overlay]

# Gradio interface
demo = gr.Interface(
    fn=segment_and_overlay,
    inputs=gr.Image(type="numpy", label="Upload Skin Image"),
    outputs=[
        gr.Image(type="numpy", label="Original Image"),
        gr.Image(type="numpy", label="Predicted Mask"),
        gr.Image(type="numpy", label="Overlay")
    ],
    title="Skin Lesion Segmentation Dashboard",
    description="Upload a skin image to visualize the predicted lesion mask using U-Net. Outputs include raw mask and overlay."
)

# Launch
demo.launch(debug=True)
```
